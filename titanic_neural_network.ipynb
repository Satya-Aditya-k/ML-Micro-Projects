```python
import numpy as np
import pandas as pd
```

```python
import tensorflow as tf
print(tf.__version__)
```

```python
train_df = pd.read_csv(r'C:\Users\adity\ML Practice folder\dataSets\titanic\train.csv')
test_df = pd.read_csv(r'C:\Users\adity\ML Practice folder\dataSets\titanic\test.csv')
```

```python
X = train_df.copy()
y = train_df['Survived']
X.pop('Survived')
```

```python
X = X.drop(columns = ['Name','Ticket','Cabin'])
```

```python
X = pd.get_dummies(X,columns=['Sex'])
```

```python
X
```

```python
X.pop('PassengerId')
X
```

```python
X['Age'] = X['Age'].fillna(30.272590)
X['Embarked'] = X['Embarked'].fillna('C')
```

```python
X = pd.get_dummies(X,columns=['Embarked'])
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
continuous = ['Age', 'Fare', 'Parch', 'Pclass', 'SibSp']
for var in continuous:
    X[var] = X[var].astype('float64')
    X[var] = scaler.fit_transform(X[var].values.reshape(-1, 1))
```

```python
model  = tf.keras.models.Sequential([
    tf.keras.layers.Dense(12,input_dim=10,activation='linear'),
    #tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(12,activation='linear'),
    tf.keras.layers.Dense(1,activation='sigmoid')
    
])
```

```python
print(model.summary())
```

```python
model.compile(optimizer='sgd',
            loss='binary_crossentropy',
             metrics=['accuracy'])
```

```python
model.fit(X, y,validation_split=0.2, epochs=100, batch_size=30)
```

```python
test_df = pd.get_dummies(test_df,columns=['Sex','Embarked'])
```
